```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Reading the CSV file
df = pd.read_csv('bdims.csv')
# Calculating the correlation between height and chest diameter
correlation = df['hgt'].corr(df['che_di'])
df['BMI'] = df['wgt'] / (df['hgt'] / 100) ** 2

plt.figure(figsize=(10, 6))
sns.regplot(x='hgt', y='che_di', data=df)
plt.title(f"Figure 1: Relationship between Height and Chest Diameter\nCorrelation: {correlation:.2f}", fontsize=14)
plt.xlabel('Height (cm)', fontsize=12)
plt.ylabel('Chest Diameter (cm)', fontsize=12)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='sex', y='BMI', data=df)
plt.title('Figure 2: Boxplot of BMI Distribution by Sex', fontsize=15)
plt.xlabel('Sex (1: Male, 0: Female)', fontsize=12)
plt.ylabel('BMI (kg/m^2)', fontsize=12)
plt.show()
```

```python
correlation2 = df['wri_di'].corr(df['ank_di'])

sns.jointplot(x='wri_di', y='ank_di', data=df)
plt.suptitle(f'Figure 3: Jointplot of Wrist Diameter and Ankle Diameter\nCorrelation: {correlation2:.2f}', fontsize=12, y=1.09)
plt.xlabel('Wrist Diameter (cm)', fontsize=12)
plt.ylabel('Ankle Diameter (cm)', fontsize=12)
plt.show()
```
